```python
# To retrieve the common first male and female names from website
url_m = "https://namecensus.com/male_names_alpha.htm"
page1 = requests.get(url_m)
top_male_first_name = pd.read_html(page1.text)

url_f = "https://namecensus.com/female_names_alpha.htm"
page2 = requests.get(url_f)
top_female_first_name = pd.read_html(page2.text)

combine = pd.concat([top_male_first_name[0], top_female_first_name[0]], ignore_index=True)
tbl_common_name = pd.DataFrame(combine["Name"].str.capitalize())
```

```python
# Can be omited or combined with Aidan's code
train_df = pd.read_csv("../data/processed/character_features_train.csv")
test_df = pd.read_csv("../data/processed/character_features_test.csv")
```

```python
# Add is_common feature to check whether the first name is common or not
is_common = train_df.first_name.isin(tbl_common_name[["Name"]].Name)
train_df = train_df.assign(is_common=is_common)

is_common = test_df.first_name.isin(tbl_common_name[["Name"]].Name)
test_df = test_df.assign(is_common=is_common)
```

```python
def get_name_length(name):
    """
    Returns the length of first name.

    Parameters:
    ------
    name: (str)
    the input name

    Returns:
    -------
    length of name: (float)

    """
    return len(name)
    assert len('Mike'), 4
    assert len('Aaron'), 5

def has_last_name(name):
    """
    Returns whether the character has a last name.
    
    Parameters:
    ------
    name: (str)
    the input name
    
    Returns:
    ------
    True/False of whether the last name exists (boolean)
    
    """
    return name == name
    
def norm(appearances, year, MAX_YR=2013):
    """
    Returns number of appearances per year.
    
    Parameters:
    ------
    appearances: (float)
    the number of appearances of the character
    
    year: (float)
    the year of its first appearance
    
    Keyword arguments:
    ------
    MAX_YR: (float)
    year boundary for the source data
    
    Returns:
    ------
    appearances_per_yr: (float)
    the number of appearances per year for each character
    
    Examples:
    ------
    norm(10, 2004) = 1
    """
    if year == year and appearances == appearances:
        appearances_per_yr = appearances/(MAX_YR - year + 1)
        return appearances_per_yr
    else:
        return 
    
train_df = train_df.assign(name_len=train_df["first_name"].apply(get_name_length))
train_df = train_df.assign(has_last_name=train_df["last_name"].apply(has_last_name))
train_df["appear_per_yr"] = train_df.apply(lambda x: norm(x['appearances'], x['year']), axis=1)

test_df = test_df.assign(name_len=test_df["first_name"].apply(get_name_length))
test_df = test_df.assign(has_last_name=test_df["last_name"].apply(has_last_name))
test_df["appear_per_yr"] = test_df.apply(lambda x: norm(x['appearances'], x['year']), axis=1)
```
